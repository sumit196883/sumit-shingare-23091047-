```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed
np.random.seed(47)

# Generate random coefficients for a 5th degree polynomial
coefficients = np.random.randn(6)

# Define the polynomial function
def polynomial(x):
    return np.polyval(coefficients, x)

# Generate x values
x_values = np.linspace(-5, 5, 1000)

# Generate y values using the polynomial function
y_values = polynomial(x_values)

# Plot the polynomial
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Polynomial', color='b')

# Find the maxima and minima
maxima_x = x_values[np.argmax(y_values)]
minima_x = x_values[np.argmin(y_values)]

# Annotate maxima and minima
plt.scatter(maxima_x, np.max(y_values), color='r', label=f'Maxima ({maxima_x:.2f}, {np.max(y_values):.2f})', zorder=5)
plt.scatter(minima_x, np.min(y_values), color='g', label=f'Minima ({minima_x:.2f}, {np.min(y_values):.2f})', zorder=5)

# Add legend
plt.legend()

# Add labels and title
plt.xlabel('x')
plt.ylabel('Polynomial value')
plt.title('Plot of a Random 5th Degree Polynomial')

# Show plot
plt.grid(True)
plt.show()
```
